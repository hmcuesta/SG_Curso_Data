```python
#Preparando Datos para Regression
y, X = dmatrices('affair ~ rate_marriage + age + yrs_married + children + \
                  religious + educ + C(occupation) + C(occupation_husb)',
                  dta, return_type="dataframe")
print X.columns
```

```python
X
```

```python
X = X.rename(columns = {'C(occupation)[T.2.0]':'occ_2',
                        'C(occupation)[T.3.0]':'occ_3',
                        'C(occupation)[T.4.0]':'occ_4',
                        'C(occupation)[T.5.0]':'occ_5',
                        'C(occupation)[T.6.0]':'occ_6',
                        'C(occupation_husb)[T.2.0]':'occ_husb_2',
                        'C(occupation_husb)[T.3.0]':'occ_husb_3',
                        'C(occupation_husb)[T.4.0]':'occ_husb_4',
                        'C(occupation_husb)[T.5.0]':'occ_husb_5',
                        'C(occupation_husb)[T.6.0]':'occ_husb_6'})
```

```python
X
```

```python
y = np.ravel(y)
```

```python
y
```

```python
#Logistic Regression
model = LogisticRegression()
model = model.fit(X, y)

#precision
model.score(X, y)
```

```python
#que porcentaje tiene aventuras
y.mean()
```

```python
#Coeficientes
pd.DataFrame(zip(X.columns, np.transpose(model.coef_)))
```

```python
#Evaluacion del modelo separamos el dataset en 2 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model2 = LogisticRegression()
model2.fit(X_train, y_train)
```

```python
#evaluando el test
predicted = model2.predict(X_test)
print predicted
```

```python
#generando las probabilidades por clase
probs = model2.predict_proba(X_test)
print probs
```

```python
# generar metricas de evaluacion
print metrics.accuracy_score(y_test, predicted)
print metrics.roc_auc_score(y_test, predicted)
```

```python
#metriz de confusion
print metrics.confusion_matrix(y_test, predicted)
print metrics.classification_report(y_test, predicted)
```

```python
#Validacion Cruzada
scores = cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=10)
print scores
print scores.mean()
```
